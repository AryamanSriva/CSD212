#Q1.c) Apply Butterworth Lowpass Filter with $D_0=60$ to Q1.a)  [3 Marks]

```python
def butterworth_lowpass_filter(shape, D0, n):
    """
    Generates Butterworth lowpass filter.

    :param shape: Shape of the filter
    :param D0: Cutoff frequency
    :param n: Order of the filter
    :return: Butterworth lowpass filter
    """
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    mask = np.zeros((rows, cols), np.float32)
    for i in range(rows):
        for j in range(cols):
            distance = np.sqrt((i - crow) ** 2 + (j - ccol) ** 2)
            mask[i, j] = 1 / (1 + (distance / D0) ** (2 * n))
    return mask

D0 = 60
n = 2  
mask = butterworth_lowpass_filter(img_a.shape, D0, n)

img_filtered_d = img_d * mask

img_filtered = np.fft.ifft2(np.fft.ifftshift(img_filtered_d))

img_filtered_mag = np.abs(img_filtered)

plt.figure(figsize=(5,5))
plt.imshow(img_filtered_mag, cmap='gray')
plt.axis('off')
plt.title('Filtered Image')
plt.show(
```

#Q1.d)Apply Gaussian Lowpass Filter with $D_0=55$ to Q1.a)  [2 Marks]

```python
def generate_gaussian_lowpass_filter(shape, cutoff_frequency):
    """
    Generates a Gaussian lowpass filter.

    :param shape: Shape of the filter
    :param cutoff_frequency: Cutoff frequency
    :return: Gaussian lowpass filter
    """
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    mask = np.zeros((rows, cols), dtype=np.float32)
    
    for i in range(rows):
        for j in range(cols):
            distance = np.sqrt((i - crow) ** 2 + (j - ccol) ** 2)
            mask[i, j] = np.exp(-(distance ** 2) / (2 * (cutoff_frequency ** 2)))
    
    return mask

cutoff_frequency = 55
img_shape = (256, 256)  
mask = generate_gaussian_lowpass_filter(img_shape, cutoff_frequency)

img_filtered_d = img_d * mask

img_filtered = np.fft.ifft2(np.fft.ifftshift(img_filtered_d))

img_filtered_mag = np.abs(img_filtered)

plt.figure(figsize=(5, 5))
plt.imshow(img_filtered_mag, cmap='gray')
plt.axis('off')
plt.title('Filtered Image')
plt.show()
```
